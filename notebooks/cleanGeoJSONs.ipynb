### Cell Towers

```python
import json
import geopandas as gpd

def clean_cell_towers_geojson(input_geojson_path, output_geojson_path):
    """
    Cleans a GeoJSON file of cell towers by retaining only essential fields.
    """
    # Load the GeoJSON file using GeoPandas
    gdf = gpd.read_file(input_geojson_path)

    # Define the essential fields to retain in the properties
    essential_fields = ['LocAdd', 'Licensee', 'LocCounty', 'LocState', 'StrucType', 'LicStatus', 'Callsign']

    # Retain only the essential fields plus the geometry
    gdf = gdf[essential_fields + ['geometry']]

    # Save the cleaned GeoJSON
    gdf.to_file(output_geojson_path, driver='GeoJSON')

# Example usage using the defined variables
clean_cell_towers_geojson(input_cell_towers, output_cell_towers)
```
