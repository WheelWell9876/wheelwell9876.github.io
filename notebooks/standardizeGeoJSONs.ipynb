## Reproject GeoJSONs to EPSG:4326

```python
# Cell 3

import warnings
import os
import sys
from io import StringIO

import warnings
warnings.filterwarnings("ignore", message="Expecting property name enclosed in double quotes: line 1 column 2 (char 1)", module="fiona.ogrext")

# Create a custom stream class to filter out specific messages
class FilteredStream:
    def __init__(self, stream):
        self.stream = stream
        self.buffer = StringIO()

    def write(self, data):
        if "WARNING:fiona.ogrext:Expecting property name enclosed in double quotes" not in data:
            self.stream.write(data)
        else:
            self.buffer.write(data)  # Capture discarded data if needed

    def flush(self):
        self.stream.flush()

# Replace sys.stderr with our filtered stream
sys.stderr = FilteredStream(sys.stderr)

# Target CRS (EPSG:4326)
target_crs = 'EPSG:4326'

# Function to reproject and save GeoJSON files
def reproject_file(input_path, output_directory, target_crs):
    try:
        # Start processing
        print(f"\nProcessing file: {os.path.basename(input_path)}")

        # Read the input file using GeoPandas
        gdf = gpd.read_file(input_path)
        print(f"  - Successfully read the file.")

        # Reproject the GeoDataFrame to the target CRS
        gdf = gdf.to_crs(target_crs)
        print(f"  - Reprojected to {target_crs}.")

        # Construct the output file path
        file_name = os.path.basename(input_path)
        output_path = os.path.join(output_directory, file_name)

        # Save the reprojected GeoDataFrame to the output path
        gdf.to_file(output_path, driver='GeoJSON')
        print(f"  - Saved reprojected file to: {output_path}")
    except Exception as e:
        print(f"  - Error processing {os.path.basename(input_path)}: {e}")

# Process each GeoJSON file
for input_path in input_paths:
    reproject_file(input_path, output_directory, target_crs)

print("\nAll datasets have been reprojected.")
```
